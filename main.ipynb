```python
import numpy as np
from graphix import Circuit
from graphix_to_hugr import convert_graphix_pattern_to_hugr
from tket.circuit.build import CircBuild, H, CX

from utils import setup_jupyter_rendering

setup_jupyter_rendering()
```

```python
# Create a quantum circuit
circuit = Circuit(2)
circuit.h(0)           # Hadamard on qubit 0
circuit.cnot(0, 1)     # CNOT from qubit 0 to 1

# Convert to MBQC pattern
pattern = circuit.transpile().pattern

# Convert to HUGR
hugr = convert_graphix_pattern_to_hugr(pattern)

print(f"Created HUGR with {len(hugr)} nodes")
```
